```python
customer_data['Review Rating'] = customer_data.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
customer_data.isna().sum()
```

```python
customer_data.columns
```

```python
customer_data.columns = customer_data.columns.str.lower()
customer_data.columns
```

```python
customer_data.columns = customer_data.columns.str.replace(' ', '_')
customer_data.columns
```

```python
customer_data = customer_data.rename(columns={'purchase_amount_(usd)': 'purchase_amount_usd'})
customer_data.head(5)
```

```python
age_labels = ['Young_adult', 'Adult', 'Middle_aged', 'Senior']
customer_data['age_group'] = pd.qcut(customer_data['age'], q= 4, labels= age_labels)
```

```python
customer_data[['age', 'age_group']].head(10)
```

```python
purchase_levels = ['Low_level_purchase', 'Mid_level_purchase', 'High_level_purchase']
customer_data['purchase_level'] = pd.qcut(customer_data['previous_purchases'], q= 3, labels= purchase_levels)
```

```python
customer_data[['previous_purchases', 'purchase_level']].head(10)
```

```python
frequently_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 120,
    'Bi-weekly': 14,
    'Annually': 360,
    'Every-three-months': 90
}
customer_data['purchase_frequency_days'] = customer_data['frequency_of_purchases'].map(frequently_mapping)
```

```python
customer_data[['frequency_of_purchases', 'purchase_frequency_days']].head(10)
```

```python
customer_data.head(10)
```

# SQL: 

```python
from sqlalchemy import create_engine

postgresql_user = 'postgres'
postgresql_password = '123456'
host = 'localhost'
port = '5432'
database = 'customer_behavior'


connection_request = f'postgresql+psycopg2://{postgresql_user}:{postgresql_password}@{host}:{port}/{database}'
engine = create_engine(connection_request)
try:
    with engine.connect() as connection:
        print('Successfully connected to the PostgreSQL database')
except Exception as ex:
    print(f'failed to connect to postgreql due to this {ex}')
```

```python
table_name = 'customer_data'
customer_data.to_sql(table_name, engine,if_exists= 'replace', index= False)
print('load the data into database successfully!')
```
